```python
import tensorflow as tf
from tensorflow import keras 
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sentences1 = [
    'i love my dog',
    'I, love my cat',
    'You love my dog!'
]

# num_words:select the 100 most frequent word
tokenizer = Tokenizer(num_words = 100)
tokenizer.fit_on_texts(sentences1)
word_index = tokenizer.word_index
print(word_index)
```

```python
# using padding to ensure all the text keep the same length

sentences2 = [
    'I love my dog',
    'I love my cat',
    'You love my dog!',
    'Do you think my dog is amazing?'
]

tokenizer = Tokenizer(num_words = 100, oov_token="<OOV>") # oov_token:
tokenizer.fit_on_texts(sentences2)
word_index = tokenizer.word_index

sequences2 = tokenizer.texts_to_sequences(sentences2)

padded = pad_sequences(sequences2, maxlen=5)
# padded = pad_sequences(sequences2, maxlen = 5 ,padding = 'post',truncating = 'post')

print("\nWord Index = " , word_index)
print("\nSequences = " , sequences2)
print("\nPadded Sequences:\n",padded)
```

```python
# Try with words that the tokenizer wasn't fit to

test_data = [
    'i really love my dog',
    'my dog loves my manatee'
]

test_seq = tokenizer.texts_to_sequences(test_data)
print("\nTest Sequence = ", test_seq)

padded = pad_sequences(test_seq, maxlen=10)
print("\nPadded Test Sequence: ")
print(padded)
```
